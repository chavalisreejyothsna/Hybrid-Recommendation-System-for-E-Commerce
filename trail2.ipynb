# Remove all images

```python
import os, shutil
folder = 'C:/Users/91949/Desktop/Project draft 1/recommendor_system/'
for filename_1 in os.listdir(folder):
    f1=folder+filename_1
#     print(filename)
    if(filename_1=='.ipynb_checkpoints' or filename_1=='Untitled.ipynb' ):
        continue
    for filename in os.listdir(f1):
        file_path = os.path.join(f1, filename)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))
#         break
#     break
```
